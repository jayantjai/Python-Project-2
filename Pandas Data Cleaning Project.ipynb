```python
# Grrouping By Aggregation

import numpy as np
import pandas as pd
data = {
    'Category' : ['A','B','A','B','A','B','A','B'],
    'Store' : ['S1','S1','S2','S2','S1','S2','S2','S1'],
    'Sales' : [100,200,150,250,120,100,200,300],
    'Quantity' : [10,15,12,18,8,20,15,25],
    'Date' : pd.date_range('2023-01-01',periods=8)
}

df4 = pd.DataFrame(data)
```

```python
df4
```

```python
a = df4.groupby('Category')
```

```python
print(a)
```

```python
for i,v in a:
    print(i)
    print(v)
```

```python
df4.groupby('Category')['Sales'].sum()
```

```python
b= df4.groupby('Store')
```

```python
for i , v in b:
    print(i)
    print(v)
```

```python
df4.groupby('Store')['Sales'].sum()
```

```python
c = df4.groupby(['Category','Store'])['Sales'].sum()
```
